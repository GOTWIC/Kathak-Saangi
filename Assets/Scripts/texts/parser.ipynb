```python
import re
```

```python
# Load and print the RTF file contents
rtf_file_path = "Kathak Saangi.rtf"

with open(rtf_file_path, 'r', encoding='utf-8') as file:
    rtf_content = file.read()


# pruning

# remove everything until and including "\margr1440"
# Remove everything up to and including "\margr1440"
match = re.search(r'\\margr1440', rtf_content)
if match:
    prune_idx = match.end()
    rtf_content = rtf_content[prune_idx:]



# line removal

lines_to_remove = [r'\pard',r'\cf3', r'\def']

lines = rtf_content.split('\n')

filtered_lines = []
for line in lines:
    should_remove = False
    for phrase in lines_to_remove:
        if phrase in line:
            should_remove = True
            break
    
    if not should_remove:
        filtered_lines.append(line)

rtf_content = '\n'.join(filtered_lines)

# phrase removal

# also add \f2, \fs48 to this list
phrases_to_remove = [r'\f0', r'\fs32', r'\up0', r'\nosupersub', r'\ulnone', r'\f1', r'\i', r'\b', r'\fs34', r'\i0', r'\b0', r'\f2', r'\fs48', r'\uc0', r'\u8232']

for phrase in phrases_to_remove:
    rtf_content = re.sub(re.escape(phrase), '', rtf_content)


# color tagging

import re

def color_tag_cf2_phrases(text):
    # Pattern for \cf2 ... \cf4
    pattern_cf4 = re.compile(r'\\cf2(.*?)(?:\\cf4)', re.DOTALL)
    # Pattern for \cf2 ... \cf5
    pattern_cf5 = re.compile(r'\\cf2(.*?)(?:\\cf5)', re.DOTALL)

    # Replace \cf2 ... \cf4
    def replacer_cf4(match):
        phrase = match.group(1)
        return f'<color=red>{phrase.strip()}</color>\\cf4'

    text = pattern_cf4.sub(replacer_cf4, text)

    # Replace \cf2 ... \cf5
    def replacer_cf5(match):
        phrase = match.group(1)
        return f'<color=red>{phrase.strip()}</color>\\cf5'

    text = pattern_cf5.sub(replacer_cf5, text)

    return text

rtf_content = color_tag_cf2_phrases(rtf_content)

# Insert a newline between '\cf4' and any tab character '\t'
rtf_content = re.sub(r'(\\cf4)([ \t]+)', r'\1\n\2', rtf_content)


# remove rtf color tags
tags_to_remove = [r'\cf1', r'\cf2', r'\cf3', r'\cf4', r'\cf5', r'\cf6']

for tag in tags_to_remove:
    rtf_content = re.sub(re.escape(tag), '', rtf_content)

# remove all '\'
rtf_content = re.sub(r'\\', '', rtf_content)

# remove starting and trailing spaces
rtf_content = rtf_content.strip()





print(rtf_content)
```
